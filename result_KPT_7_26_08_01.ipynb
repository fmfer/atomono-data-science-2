```python
import pandas as pd
import MeCab
import ipadic

CHASEN_ARGS = r' -F "%m\t%f[7]\t%f[6]\t%F-[0,1,2,3]\t%f[4]\t%f[5]\n"'
CHASEN_ARGS += r' -U "%m\t%m\t%m\t%F-[0,1,2,3]\t\t\n"'

# データをdfに読み込み
df = pd.read_csv("material/KPT_7_26_08_01/Jira.csv")

# 必要な列だけを含む新しいDataFrameを作成する
new_df = df[["Summary", "Reporter","Description", "Comment"]].copy()

# MeCabのインスタンスを作成
#m = MeCab.Tagger("-Owakati")
m = MeCab.Tagger(ipadic.MECAB_ARGS + CHASEN_ARGS)


# 文章数と単語数をカウントする関数を定義
def count_sentences_and_words(text):
    if not isinstance(text, str):
        return pd.Series([0, 0])
    sentences = text.count('。') + text.count('！') + text.count('？') + text.count('\n') + text.count(',') + 1

    words = len(m.parse(text).splitlines()[:-1])
    return pd.Series([sentences, words])

# 各列に関数を適用して新しい列を作成
for col in ["Description", "Comment"]:
    new_df[[f'{col}_sentences', f'{col}_words']] = new_df[col].apply(count_sentences_and_words)

# レポーターごとに文章数と単語数を合計
reporter_counts = new_df.groupby('Reporter').sum()

# 各レポーターの単語数と文章数の合計を計算
reporter_counts['total_sentences'] = reporter_counts[[f'{col}_sentences' for col in ["Description", "Comment"]]].sum(axis=1)
reporter_counts['total_words'] = reporter_counts[[f'{col}_words' for col in ["Description", "Comment"]]].sum(axis=1)

# 結果を表示
print(reporter_counts[['total_sentences', 'total_words']])
```

```python
import pandas as pd
import MeCab
import ipadic
from collections import Counter

CHASEN_ARGS = r' -F "%m\t%f[7]\t%f[6]\t%F-[0,1,2,3]\t%f[4]\t%f[5]\n"'
CHASEN_ARGS += r' -U "%m\t%m\t%m\t%F-[0,1,2,3]\t\t\n"'

# データをdfに読み込み
df = pd.read_csv("material/KPT_7_26_08_01/Jira.csv")

# 必要な列だけを含む新しいDataFrameを作成する
new_df = df[["Summary", "Reporter","Description", "Comment"]].copy()

# MeCabのインスタンスを作成
m = MeCab.Tagger(ipadic.MECAB_ARGS + CHASEN_ARGS)

# 単語をカウントする関数を定義
def count_words(text):
    if not isinstance(text, str):
        return []
    return m.parse(text).splitlines()[:-1]

# 各列に関数を適用して新しい列を作成
for col in ["Description", "Comment"]:
    new_df[f'{col}_words'] = new_df[col].apply(count_words)

# レポーターごとに単語を集計
reporter_words = new_df.groupby('Reporter').sum()

# 各レポーターの最も頻繁に出現する単語の上位3つを計算
for reporter, row in reporter_words.iterrows():
    words = [word for sublist in row if isinstance(sublist, list) for word in sublist if not (word.split('\t')[3].startswith('記号') or word.split('\t')[3].startswith('助詞') or word.split('\t')[3].startswith('助動詞'))]  # 記号、助詞、助動詞を除外
    word_counts = Counter(words)
    top_three = word_counts.most_common(10)
    print(f'Reporter: {reporter}')
    for word, count in top_three:
        print(f'    Word: {word}, Count: {count}')
```

#### Slack

```python
import os
import json
import MeCab
import ipadic
from collections import Counter
import re

CHASEN_ARGS = r' -F "%m\t%f[7]\t%f[6]\t%F-[0,1,2,3]\t%f[4]\t%f[5]\n"'
CHASEN_ARGS += r' -U "%m\t%m\t%m\t%F-[0,1,2,3]\t\t\n"'

# JSONファイルが格納されているディレクトリのパス
directory = 'material/KPT_7_26_08_01/Slack/'

# MeCabのインスタンスを作成
m = MeCab.Tagger(ipadic.MECAB_ARGS + CHASEN_ARGS)

# 単語のカウントを格納するCounterオブジェクト
word_counts = Counter()

# ディレクトリとそのサブディレクトリ内のすべてのファイルを読み込む
for subdir, dirs, files in os.walk(directory):
    for filename in files:
        if filename.endswith(".json"):
            filepath = subdir + os.sep + filename
            with open(filepath, 'r') as f:
                data = json.load(f)
                for message in data:
                    if 'text' in message:
                        text = message['text']
                        # HTMLエンティティを除外
                        text = re.sub(r'&[a-z]+;', '', text)
                        parsed_text = m.parse(text).split('\n')
                        words = [word.split('\t')[0] for word in parsed_text if word.strip() and len(word.split('\t')) > 3 and (word.split('\t')[3].startswith('名詞') or word.split('\t')[3].startswith('動詞') or word.split('\t')[3].startswith('形容詞')) and len(word.split('\t')[0]) > 1]  # 名詞、動詞、形容詞のみを抽出し、一文字の単語や英字のみの単語を除外
                        word_counts.update(words)

# 最も頻繁に出現する単語の上位3つを出力
top_three = word_counts.most_common(10)
for word, count in top_three:
    print(f'Word: {word}, Count: {count}')
```

```python
import MeCab
import ipadic
from collections import Counter
import re

CHASEN_ARGS = r' -F "%m\t%f[7]\t%f[6]\t%F-[0,1,2,3]\t%f[4]\t%f[5]\n"'
CHASEN_ARGS += r' -U "%m\t%m\t%m\t%F-[0,1,2,3]\t\t\n"'

# マークダウンファイルのパス
filepath = 'material/KPT_7_26_08_01/Notion.md'

# MeCabのインスタンスを作成
m = MeCab.Tagger(ipadic.MECAB_ARGS + CHASEN_ARGS)

# 単語のカウントを格納するCounterオブジェクト
word_counts = Counter()

# ファイルを開き、内容を読み込む
with open(filepath, 'r') as f:
    content = f.read()

# テキストを解析し、単語を抽出
parsed_text = m.parse(content).split('\n')
words = [word.split('\t')[0] for word in parsed_text if word.strip() and len(word.split('\t')) > 3 and len(word.split('\t')[0]) > 1 and word.split('\t')[0].isalpha() and re.match(r'^[ぁ-ん]+$', word.split('\t')[0]) and not word.split('\t')[3].startswith('助詞')]  # ひらがなのみの単語を抽出し、助詞を除外

# 単語のカウントを更新
word_counts.update(words)

# 最も頻繁に出現する単語の上位3つを出力
top_three = word_counts.most_common(10)
for word, count in top_three:
    print(f'Word: {word}, Count: {count}')
```
